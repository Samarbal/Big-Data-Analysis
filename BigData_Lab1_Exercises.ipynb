```python
# import labraries
import numpy as np
import pandas as pd
import random
import time
```

```python
def generate_sales_data(n_rows):
    # 1. Define possible values
    regions = ['Palestine', 'Syria', 'Lebanon', 'Jorden', 'Egypt', 'Qatar', 'Sudia Arabia', 'Emarates', 'Omman', 'Kwait']
    categories = ['Electronics', 'Home Decor', 'Apparel', 'Grocery', 'Health', 'Clothes', 'Axccosries']

    # 2. Generate unique data
    data = {
        'date': pd.to_datetime(np.random.choice(pd.date_range('2023-01-01', '2025-12-31'), n_rows)),
        'customer_id': np.random.randint(10000, 99999, n_rows),
        'product_category': np.random.choice(categories, n_rows),
        'amount': np.random.uniform(10.0, 500.0, n_rows).round(2),
        'region': np.random.choice(regions, n_rows)
    }

    df = pd.DataFrame(data)

    # Inject Missing Values (5%)
    for col in ['amount', 'product_category', 'region']:
        mask = np.random.rand(n_rows) < 0.05
        df.loc[mask, col] = np.nan

    # Inject Duplicates (2%)
    n_dupes = int(n_rows * 0.02)
    dupe_indices = np.random.choice(df.index, size=n_dupes)
    df_dupes = df.loc[dupe_indices]
    df = pd.concat([df, df_dupes], ignore_index=True)

    return df
```

```python
# display 5 row of the dataset
dataset = generate_sales_data(10000)
display(dataset.head())
```

```python
# data quality report  *******************
missing_values = dataset.isnull().sum()
duplicate_count = dataset.duplicated().sum()

print("Data Quality Report:")
print("---------------------------------")
print(f"Missing Values:\n{missing_values}")
print("---------------------------------")
print(f"Duplicate Rows: \n {duplicate_count}")
print("---------------------------------")
```

```python
# Memory Usage
memory_bytes = dataset.memory_usage(deep=True).sum()
memory_mb = memory_bytes / (1024**2)

print(f"Memory Usage: {memory_mb} MB ")
```

```python
# Processing time for aggregations
start_time = time.time()
sales_by_region = dataset.groupby('region')['amount'].sum().reset_index()
sales_by_region = sales_by_region.sort_values(by='amount', ascending=False)
end_time = time.time()
processing_time = end_time - start_time
print ("---------------------------------")
print(f"Processing Time for Aggregations: {processing_time} seconds")
print ("---------------------------------")
display(sales_by_region)
```

```python
def recommend_tool(rows, columns, update_frequency):
    """
    rows: number of rows
    columns: number of columns
    update_frequency: 'batch' or 'real-time'

    Returns: recommended tool and reason
    """

    # 1. Real-time always highest priority
    if update_frequency == 'real-time':
        return "Spark Streaming", "Handles real-time data processing efficiently"

    # 2. Estimate data size roughly (rows * columns)
    data_size = rows * columns

    # 3. Decision logic
    if data_size < 100_000:
        return "Excel/CSV", "Small dataset, easy to process locally"

    elif data_size < 50_000_000:
        return "Pandas (Python)", "Medium dataset, fits in memory"

    else:
        return "Spark", "Large dataset, requires distributed computing"


```

```python
# Test

# case 1 : small dataset
print(f"Case 1: {recommend_tool(5000, 10, 'batch')}")

# case 2: big dataset
print(f"Case 2: {recommend_tool(5000000, 50, 'batch')}")

# case 3: large number of columns ( medium)
print(f"Case 3: {recommend_tool(10000, 30, 'batch')}")

# case 4: many columns but fewer rows
print(f"Case 4: {recommend_tool(10000, 500, 'batch')}")

# case 5: real time processing
print(f"Case 5: {recommend_tool(10000, 10, 'real-time')}")
```
